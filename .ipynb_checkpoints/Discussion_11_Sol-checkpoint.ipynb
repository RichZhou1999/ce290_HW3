````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.table import Table
````

````python
grid_size = 4
````

````python
# left, top, right, bottom
actions = [np.array([0, -1]),
           np.array([-1, 0]),
           np.array([0, 1]),
           np.array([1, 0])]
ACTIONS_FIGS=[ '←', '↑', '→', '↓']
````

````python
actions_prob = 1/grid_size
````

````python
def is_terminal(state):
    x, y = state
    return (x == 0 and y == 0) or (x == grid_size - 1 and y == grid_size - 1)
````

````python
def step(state, action):
    if is_terminal(state):
        return state, 0

    next_state = (np.array(state) + action).tolist()
    x, y = next_state

    if x < 0 or x >= grid_size or y < 0 or y >= grid_size:
        next_state = state

    reward = -1
    return next_state, reward
````

````python
def compute_state_value(max_iter=9, discount=1.0, policy = actions_prob*np.ones((grid_size,grid_size,4))):
    new_state_values = np.zeros((grid_size, grid_size))
    iteration = 0
    
    
    while iteration<=max_iter:
      state_values = new_state_values.copy()
      old_state_values = state_values.copy()

      for i in range(grid_size):
          for j in range(grid_size):
              value = 0
              for k,a in enumerate(actions):
                  (next_i, next_j), reward = step([i, j], a)
                  value += policy[i,j,k] * (reward + discount * state_values[next_i, next_j])
              new_state_values[i, j] = value

      iteration += 1 
      
    return new_state_values, iteration
````

````python
def greedy_Policy(values,discount = 1):
    new_state_values = values
    policy = np.zeros((grid_size,grid_size,4))

    state_values = new_state_values.copy()

    for i in range(grid_size):
        for j in range(grid_size):
            value = np.min(values);
            for k,a in enumerate(actions):
                (next_i, next_j), reward = step([i, j], a)
                valtemp = reward + discount * state_values[next_i, next_j]
                if valtemp > value:
                  value = valtemp
                  actionind = k

            policy[i,j,actionind] = 1

    return policy
````

````python
def draw_policy(optimal_values,policy):
    fig, ax = plt.subplots()
    ax.set_axis_off()
    tb = Table(ax, bbox=[0, 0, 1, 1])

    nrows, ncols = optimal_values.shape
    width, height = 1.0 / ncols, 1.0 / nrows

    # Add cells
    for (i, j), val in np.ndenumerate(optimal_values):
        val=''
        for k in range(4):
            if policy[i,j,k] > 0:
              val+=ACTIONS_FIGS[k]
        
            tb.add_cell(i, j, width, height, text=val,
                loc='center', facecolor='white')

    # Row and column labels...
    for i in range(len(optimal_values)):
        tb.add_cell(i, -1, width, height, text=i+1, loc='right',
                    edgecolor='none', facecolor='none')
        tb.add_cell(-1, i, width, height/2, text=i+1, loc='center',
                   edgecolor='none', facecolor='none')

    ax.add_table(tb)
````

````python
def plot_value(values):
  fig, ax = plt.subplots()
  ax.set_axis_off()
  tb = Table(ax, bbox=[0, 0, 1, 1])

  nrows, ncols = values.shape
  width, height = 1.0 / ncols, 1.0 / nrows
  for (i, j), val in np.ndenumerate(np.round(values, decimals=4)):
      tb.add_cell(i, j, width, height, text=val,
                    loc='center', facecolor='white')

    # Row and column labels...
  for i in range(len(values)):
      tb.add_cell(i, -1, width, height, text=i+1, loc='right',
                    edgecolor='none', facecolor='none')
      tb.add_cell(-1, i, width, height/2, text=i+1, loc='center',
                    edgecolor='none', facecolor='none')
  ax.add_table(tb)
  plt.show()
````

````python
values, sync_iteration = compute_state_value(max_iter=20)
plot_value(values)
````

````python
policy0 = actions_prob*np.ones((grid_size,grid_size,4))
values, sync_iteration = compute_state_value(max_iter=20, policy = policy0)
plot_value(values)
draw_policy(values,policy0)
greedy_policy = greedy_Policy(values)
values, sync_iteration = compute_state_value(max_iter=20, policy = greedy_policy)
plot_value(values)
draw_policy(values,greedy_policy)
````
